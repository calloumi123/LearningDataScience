```python
import string
def basic_feature_extraction_setup(df):
       # Process raw text
    df['doc_object_raw'] = df['text'].apply(nlp)
    
    # Remove punctuation and newlines
    df['text_clean'] = df['text'].str.translate(str.maketrans('', '', string.punctuation)).replace('\n', ' ')
    
    # Process untidy text
    df['doc_object_untidy'] = df['text_clean'].apply(nlp)
    
    # Remove extra whitespace
    df['text_tidy'] = df['text_clean'].str.replace(r'\s+', ' ', regex=True).str.strip()
    
    # Process tidy text
    df['doc_object_tidy'] = df['text_tidy'].apply(nlp)
    
    # Remove stopwords and punctuation
    df['no_stopwords'] = [[token.text for token in doc if not token.is_stop and not token.is_punct] for doc in df['doc_object_tidy']]
    
    return df
```

```python
def avg_word_length(doc):
    words = [token.text for token in doc if not token.is_punct and not token.is_space]
    return sum(len(word) for word in words) / len(words) if words else 0
```

```python
def total_sentence_count(doc):
    return len(list(doc.sents))
```

```python
def average_sentence_length(doc):
    sentences = list(doc.sents)
    return sum(len(sentence) for sentence in sentences) / len(sentences) if sentences else 0
```

```python
from collections import Counter
def pos_distribution(doc):
    pos_counts = Counter(token.pos_ for token in doc if not token.is_punct and not token.is_space)
    return dict(pos_counts)
```

```python
def named_entity_count(doc):
    return len(doc.ents)
```

```python
def average_noun_length(doc):
    nouns = [token for token in doc if token.pos_ == 'NOUN']
    return sum(len(token.text) for token in nouns) / len(nouns) if nouns else 0
```

```python
def lexical_diversity(doc):
    words = [token.text.lower() for token in doc if not token.is_punct and not token.is_space]
    return len(set(words)) / len(words) if words else 0
```

```python
import syllapy

def average_syllables_per_word(doc):
    words = [token.text for token in doc if not token.is_punct and not token.is_space]
    syllable_counts = sum(syllapy.count(word) for word in words)
    return syllable_counts / len(words) if words else 0
```

```python
def longest_word_length(doc):
    words = [token.text for token in doc if not token.is_punct and not token.is_space]
    return max(len(word) for word in words) if words else 0
```

```python
def word_frequency_distribution(doc):
    words = [token.text.lower() for token in doc if not token.is_punct and not token.is_space]
    return dict(Counter(words).most_common(10))  # Top 10 most common words
```

```python
def extract_features(df):
    doc = df['doc_object_tidy'].iloc[1]
    docraw = df['doc_object_raw'].iloc[1]
    docstop = df['no_stopwords'].iloc[1]
    
    features = {
        'Total Sentence Count': total_sentence_count(docraw),
        'Average Sentence Length': average_sentence_length(docraw),
        'POS Distribution': pos_distribution(doc),
        'Named Entity Count': named_entity_count(doc),
        'Average Noun Length': average_noun_length(doc),
        'Lexical Diversity': lexical_diversity(doc),
        'Lexical Diversity on Meaningfull Word': lexical_diversity(docstop),
        'Average Syllables per Word': average_syllables_per_word(doc),
        'Average Syllables per Meaningfull Word': average_syllables_per_word(docstop),
        'Longest Word Length': longest_word_length(doc),
        'Word Frequency Distribution': word_frequency_distribution(docstop)
    }
    return features
```

```python
def createdatafordraw (DF):
    DFfeatures = DF[['file_location', 'page','no_stopwords']].copy()

    DFfeatures['Total_Sentence_Count'] = DF['doc_object_raw'].apply(total_sentence_count)
    DFfeatures['Average_Sentence_Length'] = DF['doc_object_raw'].apply(average_sentence_length)
    DFfeatures['POS_Distribution'] = DF['doc_object_tidy'].apply(pos_distribution)
    DFfeatures['Named_Entity_Count'] = DF['doc_object_tidy'].apply(named_entity_count)
    DFfeatures['Average_Noun_Length'] = DF['doc_object_tidy'].apply(average_noun_length)
    DFfeatures['Lexical_Diversity'] = DF['doc_object_tidy'].apply(lexical_diversity)
    DFfeatures['Meaningfull_Words_Lexical_Diversity'] = DF['no_stopwords'].apply(lexical_diversity)
    DFfeatures['Average_Syllables'] = DF['no_stopwords'].apply(average_syllables_per_word)
    DFfeatures['Meaningfull_Words_Average_Syllables'] = DF['no_stopwords'].apply(average_syllables_per_word)
    DFfeatures['Longest_Word_Length'] = DF['doc_object_tidy'].apply(longest_word_length)
    DFfeatures['Meaningfull_Words_Frequency_Distribution'] = DF['no_stopwords'].apply(word_frequency_distribution)
    return DFfeatures
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def plot_numeric_columns(df1, df2, type1_name='SL', type2_name='Other'):
    # Add Type column to each DataFrame
    df1['Type'] = type1_name
    df2['Type'] = type2_name

    # Combine the DataFrames
    combined_df = pd.concat([df1, df2], ignore_index=True)
    combined_df['Dummy'] = 1

    # Get numeric columns
    numeric_columns = combined_df.select_dtypes(include=[np.number]).columns.tolist()
    numeric_columns.remove('Dummy')  # Remove the Dummy column from the list

    # Create plots for each numeric column
    for column in numeric_columns:
        plt.figure(figsize=(12, 6))
        sns.stripplot(x='Dummy', y=column, hue='Type', data=combined_df, 
                      jitter=True, dodge=True, alpha=0.6)

        # Customize the plot
        plt.title(f'Comparison of {column} between {type1_name} and {type2_name}')
        plt.xlabel('')  # Remove x-axis label
        plt.ylabel(column)

        # Remove x-axis ticks
        plt.xticks([])

        # Show the plot
        plt.show()
    return combined_df

# Usage
combined_df = plot_numeric_columns(DFfeaturesLet, DFfeaturesRan)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import seaborn as sns
from collections import Counter
import spacy

# Load spaCy model (make sure you have a model with word vectors)
nlp = spacy.load("en_core_web_lg")  # or "en_core_web_lg" for better, but larger, word vectors

def embed_and_cluster_meaningful_words(df, word_col, type_col, n_clusters=2):
    # Get all unique meaningful words from the dataframe
    all_words = set()
    for doc in df[word_col]:
        if isinstance(doc, spacy.tokens.doc.Doc):
            all_words.update(token.text for token in doc if token.has_vector)
        elif isinstance(doc, list):
            all_words.update(word for word in doc if nlp(word).has_vector)
        else:
            raise ValueError(f"Unexpected type in {word_col} column: {type(doc)}")
    
    # Get embeddings for valid words
    word_embeddings = {word: nlp(word).vector for word in all_words if nlp(word).has_vector}
    
    if not word_embeddings:
        raise ValueError("No valid word embeddings found. Check your input data.")
    
    # Convert to numpy arrays for TSNE and KMeans
    words = np.array(list(word_embeddings.keys()))
    vectors = np.array(list(word_embeddings.values()))
    
    print(f"Shape of vectors: {vectors.shape}")  # Debugging line
    
    if vectors.shape[0] < 2:
        raise ValueError(f"Not enough valid vectors. Shape: {vectors.shape}")
    
    # Perform dimensionality reduction
    tsne = TSNE(n_components=2, random_state=42)
    word_tsne = tsne.fit_transform(vectors)
    
    # Perform clustering
    kmeans = KMeans(n_clusters=min(n_clusters, vectors.shape[0]), random_state=42)
    clusters = kmeans.fit_predict(vectors)
    
    # Create a dataframe for plotting
    plot_df = pd.DataFrame({
        'word': words,
        'x': word_tsne[:, 0],
        'y': word_tsne[:, 1],
        'cluster': clusters
    })
    
    # Calculate word importance for each document type
    for doc_type in df[type_col].unique():
        type_docs = df[df[type_col] == doc_type]
        type_words = Counter()
        for doc in type_docs[word_col]:
            if isinstance(doc, spacy.tokens.doc.Doc):
                type_words.update(token.text for token in doc if token.has_vector)
            elif isinstance(doc, list):
                type_words.update(word for word in doc if nlp(word).has_vector)
        total_words = sum(type_words.values())
        plot_df[f'importance_{doc_type}'] = plot_df['word'].map(lambda w: type_words.get(w, 0) / total_words if total_words else 0)
    
    # Plot
    plt.figure(figsize=(15, 10))
    for cluster in range(min(n_clusters, vectors.shape[0])):
        cluster_words = plot_df[plot_df['cluster'] == cluster]
        plt.scatter(cluster_words['x'], cluster_words['y'], label=f'Cluster {cluster}')
    
    # Annotate words
    for _, row in plot_df.iterrows():
        plt.annotate(row['word'], (row['x'], row['y']), 
                     xytext=(5, 5), textcoords='offset points', 
                     fontsize=8, alpha=0.7)
    
    plt.title("Meaningful Word Clusters based on spaCy Embeddings", fontsize=16)
    plt.legend()
    plt.axis('off')
    plt.tight_layout()
    plt.show()
    
    return plot_df

# Assuming your DataFrame is called 'df'
# 'meaningful_words' is the column with spaCy Doc objects containing meaningful words
# 'document_type' is the column with document types
result_df = embed_and_cluster_meaningful_words(combined_df, 'no_stopwords', 'Type')

# Now let's create a heatmap to show word importance across document types
plt.figure(figsize=(15, 10))
heatmap_data = result_df[[col for col in result_df.columns if col.startswith('importance_')]]
sns.heatmap(heatmap_data, cmap="YlOrRd", annot=True, fmt=".3f")
plt.title("Meaningful Word Importance Across Document Types", fontsize=16)
plt.xlabel("Document Type", fontsize=12)
plt.ylabel("Meaningful Words", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction import DictVectorizer
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
def visualize_document_types(df, word_freq_column, doc_type_column):
    # Extract word frequency dictionaries
    word_freq_dicts = df[word_freq_column].tolist()
    
    # Use DictVectorizer to convert dictionaries to a feature matrix
    vectorizer = DictVectorizer(sparse=False)
    word_freq_matrix = vectorizer.fit_transform(word_freq_dicts)
    
    # Perform t-SNE
    tsne = TSNE(n_components=2, random_state=42, perplexity=70)
    tsne_results = tsne.fit_transform(word_freq_matrix)
    
    # Create a new DataFrame with t-SNE results
    tsne_df = pd.DataFrame(data=tsne_results, columns=['TSNE1', 'TSNE2'])
    tsne_df['Document Type'] = df[doc_type_column]
    
    # Set up the plot
    plt.figure(figsize=(12, 8))
    sns.scatterplot(data=tsne_df, x='TSNE1', y='TSNE2', hue='Document Type', palette='deep')
    
    plt.title('Document Types Visualized by Word Frequencies')
    plt.xlabel('t-SNE Dimension 1')
    plt.ylabel('t-SNE Dimension 2')
    
    # Add a legend
    plt.legend(title='Document Type', loc='best')
    
    # Show the plot
    plt.show()
visualize_document_types(combined_df,'Meaningfull_Words_Frequency_Distribution', 'Type')
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_document_types(df, word_freq_column, doc_type_column, use_random_init=True, max_features=1000):
    # Convert frequency dictionaries to weighted strings
    def dict_to_weighted_string(freq_dict):
        return ' '.join([word + ' ' * int(freq) for word, freq in freq_dict.items()])
    
    weighted_docs = df[word_freq_column].apply(dict_to_weighted_string)
    
    # Use TfidfVectorizer to convert weighted strings to a feature matrix
    vectorizer = TfidfVectorizer(token_pattern=r'\S+', max_features=max_features)
    word_freq_matrix = vectorizer.fit_transform(weighted_docs).toarray()  # Convert to dense array
    
    # Perform t-SNE
    tsne = TSNE(n_components=2, random_state=42, init='random' if use_random_init else 'pca')
    tsne_results = tsne.fit_transform(word_freq_matrix)
    
    # Create a new DataFrame with t-SNE results
    tsne_df = pd.DataFrame(data=tsne_results, columns=['TSNE1', 'TSNE2'])
    tsne_df['Document Type'] = df[doc_type_column]
    
    # Set up the plot
    plt.figure(figsize=(12, 8))
    sns.scatterplot(data=tsne_df, x='TSNE1', y='TSNE2', hue='Document Type', palette='deep')
    
    plt.title('Document Types Visualized by Weighted Word Frequencies')
    plt.xlabel('t-SNE Dimension 1')
    plt.ylabel('t-SNE Dimension 2')
    
    # Add a legend
    plt.legend(title='Document Type', loc='best')
    
    # Show the plot
    plt.show()

visualize_document_types(combined_df,'Meaningfull_Words_Frequency_Distribution', 'Type')
```

```python
import torch
from transformers import AutoTokenizer, AutoModel
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

def get_model_features(text, model_name):
    # Load pre-trained model and tokenizer
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModel.from_pretrained(model_name, output_hidden_states=True)

    # Tokenize and encode the text
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512, padding=True)

    # Get model output
    with torch.no_grad():
        outputs = model(**inputs)

    # Get the hidden states
    hidden_states = outputs.hidden_states

    # Concatenate the last four hidden states
    token_embeddings = torch.cat([hidden_states[-1], hidden_states[-2], 
                                  hidden_states[-3], hidden_states[-4]], dim=-1)

    # Calculate the mean of all token embeddings to get a single vector for the sentence
    sentence_embedding = torch.mean(token_embeddings, dim=1).squeeze()

    return sentence_embedding.numpy()
```
